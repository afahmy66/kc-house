```python
for col in df.columns:
    df[col].plot.box(patch_artist = True,vert =False)
    plt.show();
```

```python
outlier_col = ["sqft_living","sqft_above","sqft_basement","sqft_living15"]


for i in df[outlier_col]:
    Q1 = df[i].quantile(q  = 0.25)
    Q3 = df[i].quantile(q = 0.75)

    IQR = Q3 -Q1

    lowerbound = Q1 - (1.5*IQR)
    upperbound = Q3 + (1.5*IQR)

    df[i] = np.where(df[i] < lowerbound,lowerbound,df[i])
    df[i] = np.where(df[i] > upperbound,upperbound,df[i])
    
```

```python
for col in df.columns:
    df[col].plot.box(patch_artist = True,vert =False)
    plt.show();
```
